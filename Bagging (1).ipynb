```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("churn_prediction_simple.csv")
```

```python
X = data.drop(columns=['customer_id','churn'])
Y = data['churn']
```

```python
## scaling

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_x = scaler.fit_transform(X)
```

```python
## splitting the dataset
from sklearn.model_selection import train_test_split as tts
xtrain, xtest, ytrain, ytest = tts(scaled_x,Y,train_size=0.8, stratify=Y, random_state = 101)
```

```python
## model buiding, bagging logistic regression

from sklearn.ensemble import BaggingClassifier as BC
classifier = BC()
classifier.fit(xtrain, ytrain)
```

```python
from sklearn.linear_model import LogisticRegression as LR

classifier = BC(base_estimator = LR(),
               n_estimators = 150,
               n_jobs = -1,
               random_state = 42)

classifier.fit(xtrain, ytrain)
predicted_values = classifier.predict(xtrain)
```

```python
from sklearn.metrics import classification_report

print(classification_report(ytrain, predicted_values))
```

```python
predicted_values = classifier.predict(xtest)
print(classification_report(ytest, predicted_values))
```

```python
## Random forest

from sklearn.ensemble import RandomForestClassifier as RFC
classifier = RFC()
```

```python
classifier.fit(xtrain, ytrain)
```

```python
print(classification_report(ytrain, predicted_values))
```

```python
predicted_values = classifier.predict(xtest)
print(classification_report(ytest, predicted_values))
```
